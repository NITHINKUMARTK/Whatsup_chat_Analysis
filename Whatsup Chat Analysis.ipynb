```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
f = open("WhatsApp Chat with AAIC JULY 2020.txt", "r", encoding = "utf-8")
```

```python
data = f.read()
print(data)
```

```python
print(type(data))
```

```python
pattern = "\d{1,2}/\d{1,2}/\d{1,2},\s\d{1,2}:\d{2}\s[ap][m]\s"
```

```python
message = re.split(pattern, data)[1:]
message
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({"dates":dates,"messages":message})
df
```

```python
type(df["dates"][0])
```

```python
df["dates"] = pd.to_datetime(df["dates"])
```

```python
df["messages"] = df["messages"].apply(lambda x:x.replace("-",""))
df
```

```python
def extract_user(item):
    users = []
    if ":" not in item:
        users.append("group notification")
    else:
        item = item.split(":")
        users.append(item[0])
    return str(users[0])
        
```

```python
df["users"] = df["messages"].apply(extract_user)
df
```

```python
def extract_messages(item):
    messages = []
    if ":" not in item:
        messages.append(item)
    else:
        item = item.split(":")
        messages.append(item[1])
    return str(messages[0])
        
```

```python
df["message"] = df["messages"].apply(extract_messages)
df
```

```python
df["year"] = df["dates"].dt.year
df["month"] = df["dates"].dt.month_name()
df["day"] = df["dates"].dt.day
df["hour"] = df["dates"].dt.hour
df["minute"] = df["dates"].dt.minute
df
```

```python
final = df[["users","message","year","month","day","hour","minute"]]
final.head()
```

# weekly and hourly trend

```python
period = []
for hour in df[["day_name","hour"]]["hour"]:
    if hour == 23:
        period.append(str(hour) + "-"+ str("00"))
    elif hour == 0:
        period.append(str("00")+ "-" +str(hour+1))
    else:
        period.append(str(hour)+ "-" + str(hour+1))
```

```python
df["period"] = period
```

```python
df
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.pivot_table(index = "day_name", columns = "period", values = "message", aggfunc = "count").fillna(0), cmap= "seismic_r")
plt.yticks(rotation = "horizontal")
plt.show()
```

# monthy trend with hours

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.pivot_table(index = "month", columns = "period", values = "messages", aggfunc = "count").fillna(0), cmap= "seismic")
plt.yticks(rotation = "horizontal")
plt.show()
```
